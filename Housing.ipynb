```python
def extract_floor(floor_value):
    try:
        floor_info = floor_value.split(' ')[0]
        if floor_info.isdigit():
            return int(floor_info)
        elif floor_info.lower() in ['ground', 'lower']:
            return 0
        else:
            return None
    except:
        return None

house['Floor'] = house['Floor'].apply(extract_floor)
```

```python
house.isna().sum()
```

```python
house.dropna(inplace=True)

```

```python
X = house.drop(columns=['Price'])
Y = house.Price
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.40, random_state=0)
```

# Applying Linear Regression

```python

column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False, handle_unknown='ignore'), ['Area Type', 'City', 'Furnishing Status']),
    remainder='passthrough'
)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(Y_test,y_pred_lr)
```


# Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(Y_test,y_pred_lasso)
```

# Applying Ridge

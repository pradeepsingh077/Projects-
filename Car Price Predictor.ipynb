# Cleaning


```python
backup = car.copy()
```

```python
car=car[car['year'].str.isnumeric()] # string operation get only numerics values
```

```python
car['year'] = car['year'].astype(int)
```

```python
car=car[car['Price']!="Ask For Price"]
```

```python
car['Price']= car['Price'].str.replace(',','').astype(int)
```

```python
car.info()
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car.info()
```

```python
car=car[~car['fuel_type'].isna()]
```

```python
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join('   ')
```

```python
car.reset_index(drop=True)
```

```python
car.info()
```

```python
car['kms_driven'] =car['kms_driven'].astype(int)
```

```python
car.describe()
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

# Model

```python
x = car.drop(columns = 'Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
columns_trans = make_column_transformer((OneHotEncoder(categories =ohe.categories_), ['name','company','fuel_type']), 
                                        remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe= make_pipeline(columns_trans, lr) #
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000): 
    x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3)
    lr = LinearRegression()
    pipe= make_pipeline(columns_trans, lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
import numpy as np
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3, random_state=np.argmax(scores))
lr = LinearRegression()
pipe= make_pipeline(columns_trans, lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test, y_pred) 
```
